### 2.1 IF

## 2.2 ELSE

```python
# aplico logica usando el elemento 'if' y 'else'
capitan = 'Carl Sagaan'
if capitan == 'Carl Sagan':
	print('Bienvenido Capitan')
else:
	print('Bienvenido Tripulante')
```

## 2.3 ELIF

```python
capitan = 'Carl Sagan y Elon Musk'
if capitan == 'Carl Sagan':
  print('Bienvenido nuevamente capitan')
elif capitan == 'Carl Sagan y Elon Musk':
  print('bienvenidos capitanes')
else: print('Bienvenidos astronautas')
```

### 3.1 FOR

```python
elementos = [1,2,3,4]
for elemento in elementos:
  # bloque de codigo a ejecutar para cada elemento de la secuencia
```

```python
# EJEMPLO CON LISTA
# declamramos la funcion

def saludoParticular(nombre_astronauta):
  print('bienvenido ' + nombre_astronauta)

astronautas = ['Elon Musk', 'Carl Sagan', 'Jeff Bezos']

for astronauta in astronautas:
  saludoParticular(astronauta)
```

```python
# for con RANGE

range(inicio, fin, paso)

# el argumento inicio es opcional y por defecto es 0. Es el primero numero de la secuencia
# el argumento fin es obligatorio. Es el numero que marca el final de la secuecia. El numero final no esta incluido en la secuncia
# el argumento paso es opcion y por defecto es 1. Es el incremento entre cada numero de la secuencia.
```

```python
for i in range(10):
  print(i)
```

```python
for i in range(0, 11, 2):
  print(i)
```

```python
# FOR CON DICCIONARIO
# keys(), values(), items()
frutas = {'manzana' : 3,
          'naranja' : 2,
          'banana' : 5
          }

for fruta in frutas.keys():
  print(fruta)
```

```python
for cantidad in frutas.values():
  print(cantidad)
```

```python
for fruta, cantidad in frutas.items():
  print(fruta  + ':' + str(cantidad))
```

```python
naves_lanzadas = 0
naves_totales = 5

while naves_lanzadas < naves_totales:
  print('nave lanzada ' + str(naves_lanzadas +1 ))
  naves_lanzadas += 1
```

```python
# ejemplo con break
naves_lanzadas = 0
naves_totales = 5

while True:
  print('nave lanzada ' + str(naves_lanzadas +1 ))
  naves_lanzadas += 1
  if naves_lanzadas == naves_totales:
    print('totasl las naves fueron lanzadas')
    break

```

### 3.4 LIST COMPREHENSION

```python
nombres = ['carl sagan','elon musk', 'jeff bezos']
def saludoParticular(nombre_astronauta):
  return ('bienvenido ' + nombre_astronauta)


nueva_lista = [saludoParticular(nombre.title()) for nombre in nombres]
print(nombres)
print(*nueva_lista, sep='\n')
```
